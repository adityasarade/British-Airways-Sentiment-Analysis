```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 10
page_size = 100

reviews = []

for i in range(1, pages + 1):

    print(f"Scraping page {i}")

    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    response = requests.get(url)

    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())
    
    print(f"   ---> {len(reviews)} total reviews")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 10  
total_pages = 25  
additional_pages = total_pages - pages  
page_size = 100

df = pd.read_csv("Cleaned_BA_reviews.csv")

reviews = []

for i in range(pages + 1, total_pages + 1): 
    print(f"Scraping page {i}")
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"
    response = requests.get(url)
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())
    print(f"   ---> {len(reviews)} reviews scraped from new pages")

new_df = pd.DataFrame({"reviews": reviews})

def clean_text(text):
    text = re.sub(r"✅ Trip Verified", "", text) 
    return text

new_df["reviews"] = new_df["reviews"].apply(clean_text)

df = pd.concat([df, new_df], ignore_index=True)

df = df.drop_duplicates(subset=["reviews"]).dropna(subset=["reviews"])

df.to_csv("Updated_Cleaned_BA_reviews.csv", index=False)

print(f"Scraping complete! Total reviews: {len(df)}. Data saved to 'Updated_Cleaned_BA_reviews.csv'.")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Vectorize reviews
vectorizer = CountVectorizer(stop_words='english')
doc_term_matrix = vectorizer.fit_transform(df['reviews'])

# Apply LDA
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(doc_term_matrix)

for i, topic in enumerate(lda.components_):
    print(f"Topic {i+1}:")
    print([vectorizer.get_feature_names_out()[index] for index in topic.argsort()[-10:]])
```

```python
# Perplexity
print(f"Model Perplexity: {lda.perplexity(doc_term_matrix)}")
```

```python
import gensim
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora import Dictionary
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora import Dictionary

corpus = [text.split() for text in df['reviews']]
id2word = Dictionary(corpus)
lda_gensim = gensim.models.LdaModel(
    corpus=[id2word.doc2bow(text) for text in corpus],
    id2word=id2word,
    num_topics=5,
    passes=10,
    random_state=42
)

coherence_model = CoherenceModel(model=lda_gensim, texts=corpus, dictionary=id2word, coherence='c_v')
coherence_score = coherence_model.get_coherence()
print(f"Coherence Score: {coherence_score}")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    text = text.lower()
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)

df['reviews_cleaned'] = df['reviews'].apply(preprocess)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

vectorizer = TfidfVectorizer(stop_words='english')
doc_term_matrix = vectorizer.fit_transform(df['reviews'])

lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(doc_term_matrix)

for i, topic in enumerate(lda.components_):
    wordcloud = WordCloud(background_color='white', width=800, height=400)

    topic_words = {vectorizer.get_feature_names_out()[index]: topic[index]
                   for index in topic.argsort()[-10:] if index < len(vectorizer.get_feature_names_out())}
    
    wordcloud.generate_from_frequencies(topic_words)

    plt.figure()
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Topic {i+1}")
    plt.show()
```

```python
from nltk.tokenize import sent_tokenize
from collections import Counter
import pandas as pd

negative_reviews = df[df['sentiment'] == 'negative']['reviews']
negative_text = " ".join(negative_reviews)

negative_sentences = sent_tokenize(negative_text)

sentence_counts = Counter(negative_sentences)

sorted_statements = sentence_counts.most_common(10)

frequent_negative_statements = pd.DataFrame(sorted_statements, columns=["Statement", "Frequency"])
print(frequent_negative_statements)
```

```python
import re
# Combine all negative reviews into one text
negative_reviews = df[df['sentiment'] == 'negative']['reviews']
negative_text = " ".join(negative_reviews)

# Regular expression to find "because" followed by up to 5 words
matches = re.findall(r'because(?:\s+\w+){1,5}', negative_text, re.IGNORECASE)

# Clean up the matches (remove "because" and strip extra spaces)
extracted_phrases = [match.replace("because", "").strip() for match in matches]

# Display the first 10 extracted phrases
print(extracted_phrases[:10])
```
